```python
org = pd.DataFrame(organization,columns=['uuid','roles','status','category_groups_list','num_funding_rounds','total_funding_usd','founded_on','last_funding_on','closed_on','employee_count','country_code','facebook_url','linkedin_url','twitter_url','short_description'])
org = org.loc[org.loc[:,'founded_on'] < '2014-01-01']
org = org.loc[org.loc[:,'founded_on'] > '2010-01-01']
org = org.dropna(subset=['roles'])
org = org.loc[org['roles'].str.contains('company')]
```

```python
org.isna().sum()/org.shape[0]
```

```python
org['country_code'].value_counts()
```

```python
fd3['uuid'] = fd3.index
org_fr = pd.merge(org, fd3, on="uuid", how='left')
org_fr = org_fr.dropna(subset=['seed'])
```

```python
org_fr['founded_on'] =pd.to_datetime(org_fr['founded_on'])
org_fr['found_to_seed'] = org_fr['angel/seed'] - org_fr['founded_on']
org_fr['found_to_seed'] = (org_fr['found_to_seed'].map(lambda x:x.days) / 365).round(2)

org_fr['last_funding_on'] =pd.to_datetime(org_fr['last_funding_on'])
org_fr['found_to_last'] = org_fr['last_funding_on'] - org_fr['founded_on']
org_fr['found_to_last'] = (org_fr['found_to_last'].map(lambda x:x.days) / 365).round(2)
```

```python
org_fr = org_fr.drop(org_fr[org_fr['found_to_seed'] < 0 ].index)
org_fr = org_fr.drop(org_fr[org_fr['seed_to_seA'] < 0 ].index)
org_fr = org_fr.drop(org_fr[org_fr['seA_to_seB'] < 0 ].index)
org_fr = org_fr.drop(org_fr[org_fr['seB_to_seC'] < 0 ].index)
org_fr = org_fr.drop(org_fr[org_fr['found_to_last'] < 0 ].index)
```

```python
org_fr
```

```python
cond4 = (org_fr['status'] == 'operating') & (org_fr['found_to_last'] > 4)
org_fr.loc[cond4]

org_fr = org_fr.drop(org_fr.loc[cond4].index)
```

```python
org_fr.loc[org_fr['status'] == 'operating']
```

```python
cond1 = (org_fr['seed_to_seA'] < 10 ) & (org_fr['seed_to_seA'] >0 )
cond2 = (org_fr['seA_to_seB'] < 10 ) & (org_fr['seA_to_seB'] >0 )
cond3 = (org_fr['seB_to_seC'] < 10 ) & (org_fr['seB_to_seC'] >0 )

final_cond = cond1 | cond2 | cond3
# 

org_fr.loc[(org_fr['status'] == 'operating') & final_cond ,'status'] = 'success'
org_fr.loc[(org_fr['status'] == 'operating') ]
```
