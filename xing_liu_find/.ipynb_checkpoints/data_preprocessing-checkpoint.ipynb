## Read the original data

```python
import pandas as pd
category_groups = pd.read_csv("data/category_groups.csv")
people_descriptions = pd.read_csv("data/people_descriptions.csv")
people = pd.read_csv("data/people.csv")
organization_descriptions = pd.read_csv("data/organization_descriptions.csv")
org_parents = pd.read_csv("data/org_parents.csv")
organizations = pd.read_csv("data/organizations.csv")
degrees = pd.read_csv("data/degrees.csv")
```

## Deal with people data

```python
people_descriptions = people_descriptions[["uuid","name","description"]]
```

```python
people = people[["uuid","name","gender","featured_job_organization_uuid","featured_job_organization_name","featured_job_title"]]
```

```python
people_all = pd.merge(people, people_descriptions, how='left', on=['uuid', 'name'])
```

```python
people_all.columns = ['p_uuid', 'p_name', 'gender', 'uuid',
       'name', 'featured_job_title', 'p_description']
```

```python
people_all
```

## deal with organization data

```python
organizations_select = organizations[["uuid","name","rank","roles","status","short_description","category_list","category_groups_list","num_funding_rounds","employee_count","founded_on"]]
```

```python
organization_descriptions = organization_descriptions[["uuid","name","description"]]
```

```python
organizatio_all = pd.merge(organizations_select, organization_descriptions, how='left', on=['uuid', 'name'])
```

```python
organizatio_all.head()
```

## deal with degree data

```python
degrees = degrees[["uuid","name","person_uuid","person_name","institution_uuid","institution_name","degree_type","subject"]]
```

```python
degrees.columns = ["d_uuid","d_name","p_uuid","p_name","ins_uuid","ins_name","degree_type","subject"]
```

```python
people_all_plus_degree = pd.merge(people_all, degrees, how='left', on=['p_uuid', 'p_name'])
```

```python
people_all_plus_degree.head(5)
```

```python
final_data = pd.merge(organizatio_all, people_all_plus_degree, how='left', on=['uuid', 'name'])
```

```python
final_data.columns
```

```python
final_data
```

```python
final_data = final_data.drop_duplicates(["uuid","name"])
```

```python
final_data.info()
```

```python
final_data = final_data.reset_index(drop=True)
```

```python
final_data.to_pickle("final.pkl")
```

```python
test_data = final_data.dropna(subset = ["description"])
```

```python
test_data = test_data.reset_index(drop=True)
test_data.to_pickle("test.pkl")
```

## baseline version

```python
test_data = pd.read_pickle("test.pkl")
```

```python
baseline_data = test_data[["uuid","name","status","founded_on","description","short_description","p_description","category_list"]]
```

- Get the ranked founded time
- Companies without founded time fill with time now

```python
from datetime import datetime
def parse_ymd(s):
    year_s, mon_s, day_s = s.split('-')
    return datetime(int(year_s), int(mon_s), int(day_s))
```

```python
baseline_data["founded_on"] = baseline_data["founded_on"].fillna(str(datetime.now()).split()[0])
```

```python
baseline_data["founded_on"] = baseline_data["founded_on"].apply(lambda x:parse_ymd(x))
```

- One hot encoding for investment status

```python
status_one_hot = pd.get_dummies(baseline_data["status"])
```

```python
baseline_data = pd.concat([baseline_data,status_one_hot],axis=1)
```

```python
baseline_data.to_pickle("baseline.pkl")
```
